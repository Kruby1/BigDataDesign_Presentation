### 11주차

process 1: 자전거 도로 입지에 대한 선행 연구 조사(methodology 중심으로)

1. 자전거 수요 예측 분석 선행 논문
- 자전거 수요 예측 모델 제안 (과거의 자전거 이용사용을 데이터로 한 것이 아닌 다른 변수를 이용해서) - 외국 논문
- 입체 공간 모델을 통한 경사도를 고려한 자전거 도로 제안 (딥러닝 활용) - 외국 논문
- 따릉이 데이터를 이용한 자전거 수요 예측 모델
  - 자전거 수요에 대해 중요한 변수: 1순위 자전거 우선도로 2순위 코로나19 발생 사례 3순위 전용 자전거 도로 4순위 날씨(최고 기온, 강수량, 평균 기온 등등)

2. 자전거 이용 패턴 분석 선행 논문
- 스웨덴과 영국 모두 공유 자전거를 출퇴근 시간대에 많이 사용
- 싱가포르의 경우 자전거 공유 수요가 주거 밀도, 상업 밀도 및 교차로 수와 관련이 있음을 밝힘
- 베이지안 회귀를 적용하여 자전거 공유 사용량과 인구 및 고용 밀도, 토지 이용, 대중교통 접근성 등의 요인 간의 연관성을 정량화

결론: 선행 논문 대부분 자전거 이용 수요량을 여러 변수들을 통해 수요를 예측하는 내용이거나, 현재 자전거 이용 패턴을 분석하는 것에 그친다. 이것을 토대로 새로운 자전거 도로의 입지를 제안하는 논문은 없다.
